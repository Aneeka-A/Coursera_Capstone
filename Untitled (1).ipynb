#  Segmenting and Clustering Neighborhoods in Toronto

This notebook is used to scrape the data that is in the table of postal codes from the wikipedia page, creating the desired dataframe which could be used for further data analysis, using Foursquare location data for detailed analysis and deriving results for the same.

## Scraping data from the Wikipedia page

```python
import pandas as pd
import numpy as np

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

# import the library we use to open URLs
import urllib.request

# open the url using urllib.request and put the HTML into the page variable
page = urllib.request.urlopen(url)
```

```python
# import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup

# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
soup.title #will return the HTML tags for the title and the content between them
```

```python
soup.title.string #gives us just the content string between the tags
```

```python
# using the 'find_all' function to bring back all instances of the 'table' tag in the HTML and store in 'all_tables' variable
all_tables=soup.find_all("table")
all_tables
```

```python
#taking out the right table that we will be using
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
#There are 3 columns in our table that we want to scrape the data from so we will set 3 empty lists (A, B and C) to store our data in.
A=[]
B=[]
C=[]

# To start with, we want to use the Beautiful Soup find_all function again and set it to look for the string tr. 
# We will then set up a FOR loop for each row within that array and set Python to loop through the rows, one by one.

# Within the loop we are going to use find_all again to search each row for <td> tags with the td string. 
# We will add all of these to a variable called cells and then check to make sure that there are 3 items in our cells array (i.e. one for each column).

# If yes, then we use the find(text=True)) option to extract the content string from within each <td> element in that row 
# and add them to the A-C lists we created at the start of this step.

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
# using pandas library
df_postal_codes = pd.DataFrame(A,columns=['Postal Code'])
df_postal_codes['Borough']=B
df_postal_codes['Neighborhood']=C
df_postal_codes
```

```python
df_postal_codes['Postal Code'] = df_postal_codes['Postal Code'].str.replace('\n','')
df_postal_codes['Borough'] = df_postal_codes['Borough'].str.replace('\n','')
df_postal_codes['Neighborhood'] = df_postal_codes['Neighborhood'].str.replace('\n','')
```

```python
df = df_postal_codes[df_postal_codes.Borough != 'Not assigned']
df
```

```python
df.shape
```

## Using the Geocoder package or the csv file to create the following dataframe

```python
gc = pd.read_csv('https://cocl.us/Geospatial_data')

df = pd.merge(df, gc, on='Postal Code')
df
```

## Generating maps to visualize the neighborhoods and how they cluster together

```python
!pip install geopy
from geopy.geocoders import Nominatim
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
!pip install folium
import folium
# create map of Toronto using latitude and longitude values
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
