## Covid-19 data Histogram

I will be using the matplotlib, matplotlib.pyplot, numpy and pandas for visualizing the world data into a histogram.



```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib.ticker import StrMethodFormatter

ax = df.hist(column='Cumulative_cases', bins=20, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)

ax = ax[0]
for x in ax:

    # Despine
    x.spines['right'].set_visible(False)
    x.spines['top'].set_visible(False)
    x.spines['left'].set_visible(False)

    # Switch off ticks
    x.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

    # Draw horizontal axis lines
    vals = x.get_yticks()
    for tick in vals:
        x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1)

    # Remove title
    x.set_title("")

    # Set x-axis label
    x.set_xlabel("No. of Cases", labelpad=20, weight='bold', size=12)

    # Set y-axis label
    x.set_ylabel("No. of Countries", labelpad=20, weight='bold', size=12)
    
    # Format y-axis label
x.yaxis.set_major_formatter(StrMethodFormatter('{x:,g}'))    
    
```

```python
delhi_geo = "http://pincode.india-server.com/states/delhi/"

# import the library we use to open URLs
import urllib.request

# open the url using urllib.request and put the HTML into the page variable
page = urllib.request.urlopen(delhi_geo)

# import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup

# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "lxml")

print(soup.prettify())
```

```python
# using the 'find_all' function to bring back all instances of the 'table' tag in the HTML and store in 'all_tables' variable
all_tables=soup.find_all("table")
all_tables
```

```python
#taking out the right table that we will be using
right_table=soup.find('table', class_='pincode-tbl')
right_table
```

```python
A=[]
B=[]
C=[]

# To start with, we want to use the Beautiful Soup â€˜find_allâ€™ function again and set it to look for the string â€˜trâ€™. 
# We will then set up a FOR loop for each row within that array and set Python to loop through the rows, one by one.

# Within the loop we are going to use find_all again to search each row for <td> tags with the â€˜tdâ€™ string. 
# We will add all of these to a variable called â€˜cellsâ€™ and then check to make sure that there are 3 items in our â€˜cellsâ€™ array (i.e. one for each column).

# If yes, then we use the find(text=True)) option to extract the content string from within each <td> element in that row 
# and add them to the A-C lists we created at the start of this step.

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
# using pandas library
delhi_postal_codes = pd.DataFrame(A,columns=['S.No.'])
delhi_postal_codes['District Name']=B
delhi_postal_codes['No.of post offices']=C
delhi_postal_codes
```
